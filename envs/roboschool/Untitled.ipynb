```python
import tensorflow as tf
import numpy as np
import gym
import roboschool
```

```python
env = gym.make("RoboschoolHumanoid-v1")
```

```python
nObs = env.observation_space.shape[0]
nAct = env.action_space.shape[0]
print("Observation space: {} | Action space: {}".format(nObs, nAct))
```

```python
maxStep = 500
maxEpisode = 100
```

```python
totalR = 0
done = False
s = env.reset()

for nStep in range(maxStep):
    env.render()
    s2, r, done, info = env.step(env.action_space.sample())
    
    totalR += r
    s = s2
    
    print("step: {}  | reward: {}  |".format(nStep, totalR))
    
    
    
```
